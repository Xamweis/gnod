```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

Retrieve an arbitrary Wikipedia page of "Python" and create a list of links on that page: url ='https://en.wikipedia.org/wiki/Python'

```python
url = "https://en.wikipedia.org/wiki/Python"
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
links = []

for link in soup.find_all('a'):
    if 'href' in link.attrs:
        if 'http' in link['href']:
            links.append(link['href'])

links
```

Display the 20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe: url = 'https://www.emsc-csem.org/Earthquake/'

```python
url = 'https://www.emsc-csem.org/Earthquake/'
```

```python
response = requests.get(url, headers={'Accept-Language':'en-US'})
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
times, lats, longs, regions = [], [], [], []

for row in soup.select('#tbody > tr')[:20]:
    time = row.select('.tabev6 a')[0].get_text().replace('\xa0', ' ')
    times.append(time)
    lat = row.select('td')[4].get_text().replace('\xa0', ' ') + row.select('td')[5].get_text().replace('\xa0', '')
    lats.append(lat)
    long = row.select('td')[6].get_text().replace('\xa0', ' ') + row.select('td')[7].get_text().replace('\xa0', '')
    longs.append(long)
    region = row.select('.tb_region')[0].get_text().replace('\xa0', '')
    regions.append(region)


quakes = pd.DataFrame({
    "Datetime":times,
    "Lat":lats,
    "Long":longs,
    "Region":regions
}, index=[x for x in range(1,21)])
quakes
```

List all language names and number of related articles in the order they appear in wikipedia.org: url = 'https://www.wikipedia.org/'

```python
url = "https://www.wikipedia.org/"
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
langs, arts = [], []
for lang in soup.select('.central-featured-lang'):
    langs.append(lang.strong.get_text())
    arts.append(lang.small.get_text().split('+')[0])

wiki = pd.DataFrame({
    'language':langs,
    'articles':arts
})
display(wiki)

# display([l.strong.get_text() + '\t' + l.small.get_text().split('+')[0] for l in soup.select('.central-featured-lang')])
```
