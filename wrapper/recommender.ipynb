```python
import pandas as pd
import numpy as np
import pickle

import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
secrets_file = open("../secrets/secrets.txt","r")
string = secrets_file.read()
secrets_file.close()
```

```python
secrets_dict={}
for line in string.split('\n'):
    if len(line) > 0:
        secrets_dict[line.split(':')[0]]=line.split(':')[1]
```

```python
#Initialize SpotiPy with user credentials
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=secrets_dict['clientid'],
                                                           client_secret=secrets_dict['clientsecret']))
```

```python
a = pd.read_csv('../songs_csv/jan.csv')
b = pd.read_csv('../songs_csv/willis.csv')

x = pd.read_csv('../songs_csv/mine.csv')

cluster_features = [
    'danceability', 'energy', 'loudness', 'speechiness', 
    'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms'
    ]
all = pd.concat([a,b,x],axis=0).drop_duplicates(subset=['id']).dropna(subset=cluster_features)

# fill preview-nans:
all['preview'] = all['preview'].fillna('No preview URL')

# shuffle the df and reset its index
all = all.sample(frac=1).reset_index(drop=True)
all
```

```python
billboard_100 = pd.read_csv('top100_billboard.csv')
billboard_100 = billboard_100.rename(columns={
    'Title': 'track_name',
    'Artist': 'artist'
})
billboard_100['preview'] = ['No preview URL' for _ in range(len(billboard_100))]
```

```python
def ask_user():
    while True:
        user_song = input('ENTER YOUR SONG!')
        if user_song == '':
            continue
        user_artist = input('ENTER THE ARTIST!')
        if user_artist == '':
            continue
        else:
            return user_song, user_artist
```

```python
def delete_from_top100(song, artist, top100:pd.DataFrame):
    song_artist = top100[top100['track_name'].str.lower() == song.lower()]['artist'].str.lower()
    if len(song_artist) < 1:
        return top100

    elif artist.lower() in song_artist.values:
        top100 = top100[~((top100['track_name'].str.lower() == song.lower())&(top100['artist'].str.lower() == artist.lower()))]
    
    return top100
```

```python
def get_song_uri(song, artist, spotifyclient:spotipy.Spotify):
    track = spotifyclient.search(q=f"remaster%20track:{song}%20artist:{artist}", type='track', limit=1)['tracks']['items']
    if len(track) > 0:
        return track[0]['uri']
    else:
        return
```

```python
def get_cluster_features(uri, spotifyclient:spotipy.Spotify):
    features = pd.json_normalize(spotifyclient.audio_features(uri))
    return features.reset_index(drop=True)
```

```python
def recommend_song_from_top100(song, artist, top100:pd.DataFrame):
    top100 = top100[~((top100['track_name'].str.lower() == song.lower()) &(top100['artist'].str.lower() == artist.lower()))]
    return top100.sample(n=1)
```

```python
def recommend_song_from_cluster(df:pd.DataFrame, clust: int, clustcol='cluster'):
    return df[df[clustcol]==clust].sample(n=1)
```

```python
def RECOMMENDER_3000(song, artist, top100, database, client, model, cluster_features, scaler):
    print('Your song was:', song, 'by', artist)
    new_top100 = delete_from_top100(song, artist, top100)
    
    if len(new_top100) != len(top100):
        rec = recommend_song_from_top100(song, artist, top100)
        print('Your song is in the TOP 100, check another one out:')

    else:
        database_scaled_features = scaler.transform(database[cluster_features])
        database = database.drop(['cluster'], axis=1, errors='ignore')
        database['cluster'] = model.predict(database_scaled_features)

        song_uri = get_song_uri(song, artist, client)
        if not song_uri:
            print('Song not found on spotify!')
            return
        features = get_cluster_features(song_uri, client)[cluster_features]
        scaled = scaler.transform(features)
        song_cluster = model.predict(scaled)[0]
        rec = recommend_song_from_cluster(database, song_cluster)
        print('You have a nice taste in music, check out this one:')

    print(rec['track_name'].values[0], 'by', rec['artist'].values[0])
    print(rec['preview'].values[0])
```

```python
from sklearn.preprocessing import StandardScaler

# filter out all the numericals
nums = all.select_dtypes(include=np.number)
# time_signature & key only few values --> categorical
nums = nums.drop(['time_signature', 'key', 'mode'], axis=1)

scaler = StandardScaler().fit(nums)
scaled = scaler.transform(nums)
```

```python
from sklearn import cluster

kmeans = cluster.KMeans(n_clusters=8)
kmeans.fit(scaled)
```

```python
song, artist = ask_user()
```

```python
RECOMMENDER_3000(song, artist, billboard_100, all, sp, kmeans, cluster_features, scaler)
```
